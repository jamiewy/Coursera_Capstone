```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
print('Libraries imported')
```

Get HTML data from the link, use beautifulsoup to parse HTML data.

```python
link='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
link_data=requests.get(link).text
soup=BeautifulSoup(link_data,'lxml')
```

Store the data in dataframe canada_df.

```python
rows=[]
columns=[]
table=soup.find(class_='wikitable')
for index, tr in enumerate(table.find_all('tr')):
    section=[]
    for td in tr.find_all(['th','td']):
        section.append(td.text.rstrip())
    
    if (index ==0):columns=section
    else: rows.append(section)

canada_df=pd.DataFrame(data=rows,columns=columns)
canada_df.head(12)
```

Clean dataframe by removing data with 'not assigned' cell in 'Borough'.

```python
canada_df=canada_df[canada_df['Borough']!='Not assigned']
canada_df.head(12)
```

Clean data by replace 'not assigned' in 'Neighbourhood' with the same cell in 'Borough'.

```python
canada_df['Neighbourhood'].replace('Not assigned',canada_df['Borough'],inplace=True)
canada_df.head(12)
```
